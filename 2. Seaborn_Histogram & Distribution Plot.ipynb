```python
#Importing libraries
import seaborn as sns
from scipy.stats import norm
import matplotlib.pyplot as plt
```

```python
#Loading the data
data = sns.load_dataset('tips')
```

```python
#View the data
data
```

```python
#Creating distribution plot
sns.distplot(data['size']);
```

```python
#Creating distribution plot
sns.distplot(data['tip']);
```

```python
#Creating distribution plot
sns.distplot(data['total_bill']);
```

```python
#Creating distribution plot with bins
sns.distplot(data['total_bill'], bins = 50);
```

```python
#Removing histogram from plot
sns.distplot(data['total_bill'], hist = False);
```

```python
#Removing distribution line from plot
sns.distplot(data['total_bill'], kde = False);
```

```python
#Showing rug in plot
sns.distplot(data['total_bill'], rug = True);
```

```python
#Showing standard normal distribution curve
sns.distplot(data['total_bill'], fit=norm);
```

```python
#Removing distribution curve and showing standard normal distribution curve
sns.distplot(data['total_bill'], fit=norm, kde=False);
```

```python
#Changing the color of the plot 
sns.distplot(data['total_bill'], color='r');
```

```python
#Changing direction of the plot as vertical
sns.distplot(data['total_bill'], vertical=True);
```

```python
#Creating distribution plot 
sns.distplot(data['total_bill'], norm_hist=True);
```

```python
#Creating distribution plot 
sns.distplot(data['total_bill'], 
             norm_hist=False);
```

```python
#Chnging / adding label on x axis
sns.distplot(data['total_bill'], 
             axlabel='Total Bill');
```

```python
#adding label on legend
sns.distplot(data['total_bill'],
             label='Total Bill')
plt.legend()
```

```python
#Creating distribution plot
sns.distplot(data['total_bill'],
             label='Total Bill')
plt.title('Histogram of Total Bill')
plt.legend()
```

```python
#Creating distribution plot
sns.set()
sns.distplot(data['total_bill'],
             label='Total Bill')
plt.title('Histogram of Total Bill')
plt.legend()
```

```python
#Creating distribution plot
plt.figure(figsize=(16,9))
sns.set()
sns.distplot(data['total_bill'],
             label='Total Bill')
plt.title('Histogram of Total Bill')
plt.legend()
```

```python
#Sorting variable total bills 
data.total_bill.sort_values()
```

```python
#Creating distribution plot
bins = [1,5,10,15,20,25,30,35,40,45,50,55]
plt.figure(figsize=(16,9))
sns.set()
sns.distplot(data['total_bill'],bins=bins)
plt.xticks(bins)
plt.title('Histogram of Total Bill');
```

```python
#Creating distribution plot
plt.figure(figsize=(16,9))
sns.set(style = 'dark')
sns.distplot(data['total_bill'],
             #Changing bar Color, Edge, Width & Opacity
             hist_kws={'color': 'mediumorchid', 
                       'edgecolor':'whitesmoke', 
                       'linewidth':2, 
                       'linestyle':'-.',
                       'alpha':0.5})
```

```python
#Creating distribution plot
plt.figure(figsize=(16,9))
sns.set()
sns.distplot(data['total_bill'],
            #Changing Bar Color, Edge, Width & Opacity
             hist_kws={'color': 'coral', 
                       'edgecolor':'k', 
                       'linewidth':1, 
                       'linestyle':'-',
                       'alpha':0.9},
            #Changing color of Distribution Curve Color, Edge, Width & Opacity
            kde_kws= {'color': 'r', 
                      'linewidth':3, 
                      'linestyle':':',
                      'alpha':0.9},
            #Adding rug
            rug=True,
            #Changing rug Color, Edge, Width & Opacity
            rug_kws={'color': 'g', 
                     'linewidth':5, 
                     'linestyle':'--',
                     'alpha':0.9})
```

```python
#Creating distribution plot
plt.figure(figsize=(16,9))
sns.set()
sns.distplot(data['total_bill'],
             #Changing Bar Color, Edge, Width & Opacity
             hist_kws={'color': 'purple', 
                       'edgecolor':'k', 
                       'linewidth':1, 
                       'linestyle':'-',
                       'alpha':0.5},
            #Removing Distribution Line
            kde= False,
            #Showing Standard Normal Distribution
            fit = norm,
            #Changing Standard Normal Distribution Color, Edge, Width & Opacity
            fit_kws={'color': 'b', 
                     'linewidth':5, 
                     'linestyle':':',
                     'alpha':1},
            #Showing rug
            rug=True,
            #Changing rug Color, Edge, Width & Opacity
            rug_kws={'color': 'b', 
                     'linewidth':5, 
                     'linestyle':'--',
                     'alpha':0.5});
```

```python
#Creating distribution plot
plt.figure(figsize=(16,9))
sns.set()
sns.distplot(data['total_bill'], 
             
             hist_kws={'color': 'purple', #Changing color of bars 
                       'edgecolor':'k',  #Changing color of edge of bars
                       'linewidth':1,  #Changing edge line width
                       'linestyle':'-', #Changing style of edge line
                       'alpha':0.5}, #Changing opacity of the bar
            
             kde= False, #Removing distribution line
            
            fit = norm, #Showing standard normal distribution
            fit_kws={'color': 'gold', #Changing the color of standard normal ditribution curve
                     'linewidth':5, #Changing the line width
                     'linestyle':'-',#Changing the style of line
                     'alpha':0.9}, #Changing the opacity of the curve
            
            rug=True,#Adding rug
            rug_kws={'color': 'gold', #Changing the rug color
                     'linewidth':5, #Changing the width of rug
                     'linestyle':'-', #Changing the style of rug
                     'alpha':0.5},  #Changing the opacity of rug
            
             label='TB'); # Legend label

plt.title('Histogrm of Restorant Total Bill', fontsize = 20);
plt.xlabel('Total Bill', fontsize = 15);
plt.legend();
```

```python
#Creating distribution plots of three variables
plt.figure(figsize=(20,12))
sns.distplot(data['total_bill'], bins = 9, label = 'total bill')
sns.distplot(data['tip'], bins = 9, label = 'tip')
sns.distplot(data['size'], bins = 9, label = 'size')
plt.legend()
```
