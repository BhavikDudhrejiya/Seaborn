```python
#Importing libraries
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Loading data
data = sns.load_dataset('tips')
data
```

```python
#Creating Bar Plot
sns.barplot(x = data.day, 
            y = data.total_bill);
```

```python
#Creating Bar Plot with categorical
sns.barplot(x = data.day, 
            y = data.total_bill, 
            hue=data.sex);
```

```python
#Creating Bar Plot
sns.barplot(x = 'day', 
            y = 'total_bill', 
            hue=data.sex, 
            data = data);
```

```python
#Changing the order of the x axis categorical var
order = ['Sun','Thur','Fri','Sat']

sns.barplot(x = 'day', 
            y = 'total_bill', 
            hue='sex', 
            data = data, 
            order=order);
```

```python
#Changing hue order
hue_order = ['Female', 'Male']

order = ['Sun','Thur','Fri','Sat']

sns.barplot(x = 'day', 
            y = 'total_bill', 
            hue='sex', 
            data = data,
            hue_order=hue_order);
```

```python
#Adding estimator mean
sns.barplot(x = 'day', 
            y = 'total_bill', 
            hue='sex', 
            data = data,
            hue_order=hue_order, 
            estimator=np.mean);
```

```python
#Adding estimator max
sns.barplot(x = 'day', 
            y = 'total_bill', 
            hue='sex', 
            data = data,
            hue_order=hue_order,
            estimator=np.max)
```

```python
#Adding confidence interval 95
sns.barplot(x = 'day', 
            y = 'total_bill', 
            hue='sex', 
            data = data,
            hue_order=hue_order, 
            ci =95);
```

```python
#Number of bootstrap iterations to use when computing confidence intervals
sns.barplot(x = 'day', 
            y = 'total_bill', 
            hue='sex', 
            data = data,
            hue_order=hue_order, 
            estimator=np.max, 
            n_boot=1000);
```

```python
 #Changing the direction of the plot
sns.barplot(y = 'day', 
            x = 'total_bill', 
            hue='sex', 
            data = data, 
            orient='h');
```

```python
 #Changing the direction of the plot
sns.barplot(x = 'total_bill', 
            y = 'size', 
            hue = 'sex', 
            data = data, 
            orient = 'h');
```

```python
#Changing color of the plot
sns.barplot(x = 'day', 
            y = 'total_bill', 
            hue = 'sex', 
            data = data, 
            color='g');
```

```python
#Changing paletter of the plot
sns.barplot(x = 'day', 
            y = 'total_bill', 
            hue = 'sex', 
            data = data, 
            palette='magma');
```

```python
#Changing saturation of the plot
sns.barplot(x = 'day', 
            y = 'total_bill', 
            hue = 'sex', 
            data = data, 
            saturation=0.3);
```

```python
#Changing error color
sns.barplot(x = 'day', 
            y = 'total_bill', 
            hue = 'sex', 
            data = data, 
            errcolor='0.9');
```

```python
#Changing error width
sns.barplot(x = 'day', 
            y = 'total_bill', 
            hue = 'sex', 
            data = data, 
            errwidth=5);
```

```python
#Changing cap size of an error
sns.barplot(x = 'day', 
            y = 'total_bill', 
            hue = 'sex', 
            data = data,
            capsize=0.2);
```

```python
#Widen the size of bar
sns.barplot(x = 'day', 
            y = 'total_bill', 
            hue = 'sex', 
            data = data,
            dodge=False);
```

```python
#Changing Opacity, line style, edge width, edge color
kwargs = {'alpha':0.9, 
          'linestyle':':', 
          'linewidth':2, 
          'edgecolor':'k'}

sns.barplot(x = 'day', 
            y = 'total_bill', 
            hue = 'sex', 
            data = data,
            **kwargs);
```

```python
#Changing Opacity, line style, edge width, edge color
kwargs = {'alpha':0.9, 
          'linestyle':':', 
          'linewidth':2, 
          'edgecolor':'k'}

sns.barplot(x = 'day', 
            y = 'total_bill', 
            hue = 'sex', 
            data = data,
            **kwargs);
```

```python
#Changing Opacity, line style, edge width, edge color
sns.barplot(x = 'day', 
            y = 'total_bill', 
            data = data, 
            alpha = 0.5, 
            linestyle = '-', 
            linewidth = 3, 
            edgecolor = 'g');
```

```python
#Using matplotlib axis
ax = sns.barplot(x = 'day', 
                 y = 'total_bill', 
                 data = data, 
                 alpha = 0.5, 
                 linestyle = '-', 
                 linewidth = 3, 
                 edgecolor = 'g')

ax.set(title = 'Barplot of Tips DataFrame',
      xlabel = 'Days',
      ylabel = 'Total Bill');
```

```python
#Creating bar plot
plt.figure(figsize=(16,9))

sns.barplot(x = 'day', 
            y = 'total_bill', 
            data = data, 
            alpha = 0.5, 
            linestyle = '-.', 
            linewidth = 3, 
            edgecolor = 'k')

plt.title('Barplot of Tips Data Frame', 
          fontsize = 15)

plt.xlabel('Days',
           fontsize = 15)

plt.ylabel('Total Bill',
           fontsize = 15)
```
